```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bins=[0,100000,300000,float('inf')]
labels=['Low','Med','High']

df['Income Band']=pd.cut(df['Estimated Income'],bins=bins,labels=labels,right=False)
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
categorical_cols=df[['BRId','GenderId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income Band']]
for col in categorical_cols:
  print(f'Value counts for {col} :')
  display(df[col].value_counts())
```

## UNIVARIATE ANALYSIS

```python
for i ,predictor in enumerate(df[['BRId','GenderId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income Band']]):
  plt.figure(i)
  sns.countplot(data=df,x=predictor)
```

## BIVARIATE ANALYSIS

```python
for i ,predictor in enumerate(df[['BRId','GenderId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income Band']]):
  plt.figure(i)
  sns.countplot(data=df,x=predictor,hue='GenderId')
```

```python
for i ,predictor in enumerate(df[['BRId','GenderId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income Band']]):
  plt.figure(i)
  sns.countplot(data=df,x=predictor,hue='Nationality')
```

```python
for col in categorical_cols:
  if col=="Occupation":
    continue
  plt.figure(figsize=(8,5))
  sns.histplot(df[col])
  plt.title('Histogram of Occupation Count')
  plt.xlabel('Occupation')
  plt.ylabel('Count')
  plt.show()
```

## NUMERICAL ANALYSIS

```python
numerical_cols=['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account']
plt.figure(figsize=(26,20))
for i,col in enumerate(numerical_cols):
  plt.subplot(4,3,i+1)
  sns.histplot(df[col],kde=True)
  plt.title(col)
plt.show()
```

## HEATMAPS

```python
numerical_cols=['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account']
correlation_matrix=df[numerical_cols].corr()
plt.figure(figsize=(15,10))
sns.heatmap(correlation_matrix,annot=True,cmap='crest',fmt='.2f')
plt.title('correlation matrix')
plt.show()
```
